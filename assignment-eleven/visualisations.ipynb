```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

# Plot comparative bar chart

```python
avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

avg_scores.plot(kind="bar", figsize=(8, 6))
plt.title("Average Scores of Male vs Female Students")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(title="Subjects")
plt.show()
```

# Word Cloud: Parental Level of Education

```python
text = " ".join(df["parental level of education"].astype(str))
font_path = matplotlib.get_data_path() + "/fonts/ttf/DejaVuSans.ttf"
wordcloud = WordCloud(width=800, height=400, background_color="white", font_path=font_path).generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="hamming")
plt.axis("off")
plt.title("Word Cloud of Parental Level of Education")
plt.show()
```
